```python
#Part 1
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1)

arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)

zeros_arr = np.zeros((3, 3))
print("Zeros Array:\n", zeros_arr)

ones_arr = np.ones((2, 4))
print("Ones Array:\n", ones_arr)

range_arr = np.arange(0, 10, 2)
print("Range Array:", range_arr)
```

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Data Type:", arr.dtype)
```

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

```python
#Part 2
import numpy as np

table = np.array([
    [5, 8, 2, 7],
    [1, 9, 4, 3],
    [6, 0, 11, 10],
    [12, 13, 14, 15]
])

print("Second row:", table[1, :])
print("Third column:", table[:, 2])
print("Sub-table R2-R3, C2-C3:\n", table[1:3, 1:3])
```

```python
import numpy as np

scores = np.array([
    [70, 80, 90],
    [60, 75, 85],
    [88, 92, 79],
    [55, 65, 72]
])

scores[:, 1] += 5
scores[0, :] = [100, 100, 100]
scores[:, 2] = 0

print("Updated scores:\n", scores)
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
v = np.array([2, 4])

print("Element-wise multiplication:\n", A * B)
print("Matrix multiplication (A @ B):\n", A @ B)
print("Broadcasting (A * v):\n", A * v)
```

```python
import numpy as np

np.random.seed(42)
matrix = np.random.randint(-5, 6, (4, 4))
print("Random Matrix:\n", matrix)

vals, counts = np.unique(matrix, return_counts=True)
print("Unique values:", vals)
print("Counts:", counts)

# Part B: Unique Rows
a = np.array([
    [1, 0, 0],
    [1, 0, 0],
    [2, 3, 4]
])
unique_rows = np.unique(a, axis=0)
print("Unique rows:\n", unique_rows)
```

```python
import numpy as np

x = np.array([2, 0, 1, 5, 4, 1, 9])
A = np.array([[3, 8, 1], [6, 2, 9]])

# argsort returns the indices that would sort an array
sorted_indices = np.argsort(x)
print("Indices of sorted x:", sorted_indices)

print("Sorted x:", np.sort(x))

print("Sorted A (axis 1):\n", np.sort(A, axis=1))

print("Sorted A (axis 0):\n", np.sort(A, axis=0))
```
